Check Year_film years.


```python
df = pd.read_csv('golden_globe_awards.csv')
df = df.dropna(how = 'all')
df.drop(df[df['year_film'] > 1950].index, inplace = True)
print(df.head(15))
print(df.shape)
```

Data Vizualization


```python
df = pd.DataFrame(award_df['win'].groupby(award_df['category']).sum())
df.plot(kind= 'bar')
plt.show()
```

```python
df = df.reset_index()
sns.barplot(x = df['win'], y = df['category'])
```

```python
award_df.category.value_counts(normalize=True)
award_df.category.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
plt.scatter(df.category,df.win)
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data = df, vars=['category','win'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.title("",fontsize=20)
sns.countplot('win',data=df)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
```
